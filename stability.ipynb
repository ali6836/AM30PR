```python
from sympy import symbols, I
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import animation
import mpmath
plt.rcParams['text.usetex'] = True
mpmath.mp.dps = 15
x=symbols('x',cls=sp.Function)
y,t,g,v=symbols('y t g v', real=True)
```

```python
xbar = sp.symbols('xbar')
xbar = 5
#g=-4
#v=-1


x=symbols('x',cls=sp.Function)
y,t=symbols('y t', real=True)
x=x(y,t)#sp.Piecewise((y,sp.Eq(t,0)),(x(y,t),True))
#x=x(t)
xy = sp.diff(x,y)#sp.Piecewise((1,sp.Eq(t,0)),(sp.diff(x,y),True))
xyy = sp.diff(xy,y)
xyyy = sp.diff(xyy,y)
xyyyy = sp.diff(xyyy,y)
R0 = 0.5 * sp.sech(x-xbar)
rho0 = R0 **2
P0 = sp.Abs(R0)
P0y = sp.diff(P0,y)
P0yy = sp.diff(P0y,y)
Q0 = -0.5 *P0yy / P0
Q0y = sp.diff(Q0,y)


term1 = -Q0y/(xy**3)
term2 = (4*Q0 + (P0y/P0)**2 )*xyy/(xy **4)
term3 = 3* (P0y/P0) * (xyy**2 /(xy**5))
term4 = - 1* (P0y/P0) * (xyyy**1 /(xy**4))
term5 = -(5/2)*((xyy**3)/(xy**6))
term6 = 2 * xyy*xyyy/(xy**5)
term7 = - 1/4 *xyyyy/xy**4
term8 = - 2*g*P0*P0y/xy**2
term9 = g*P0**2 *xyy/xy**3

results = term1 + term2 + term3 + term4 +term5 + term6 +term7 +term8 + term9 
#results
```

```python
k = sp.diff(results,t).subs(x,v*t +y).doit()



k2 = results.subs(x,v*t +y).doit()
h=sp.lambdify([y,t,v,g],k.doit(),'numpy')
h2=sp.lambdify([y,t,v,g],k2.doit(),'numpy')
```

```python
def func(x):
    return h2(7,x,-1,-3)
def bisection(a,b):
 
    if (func(a) * func(b) >= 0):
        print("You have not assumed right a and b\n")
        return
  
    c = a
    while ((b-a) >= 1e-5):
 
        # Find middle point
        c = (a+b)/2
  
        # Check if middle point is root
        if (func(c) == 0.0):
            break
  
        # Decide the side to repeat the steps
        if (func(c)*func(a) < 0):
            b = c
        else:
            a = c
             
    return np.float64("%.4f"%c)
     
# Driver code
# Initial values assumed
sols = []
sols.append(bisection(0, 1))
sols.append(bisection(1, 2.1))
sols.append(bisection(2.1, 5))
```
